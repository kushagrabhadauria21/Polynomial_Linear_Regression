##**importing libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

##**importing dataset**

```python
df=pd.read_csv("Position_Salaries.csv")
X=df.iloc[:,1:-1].values
y=df.iloc[:, -1].values
y
```

##**importing linear Regression**

```python
from sklearn.linear_model import LinearRegression
lin_regressor= LinearRegression()
lin_regressor.fit(X,y)
```

##**importing Polynomial Regression**

```python
from sklearn.preprocessing import PolynomialFeatures
poly_regressor= PolynomialFeatures(degree=4)
X_poly= poly_regressor.fit_transform(X)
lin_reg2=LinearRegression()
lin_reg2.fit(X_poly, y)
lin_reg2
```

##**plotting prediction of linear regression**


```python
plt.scatter(X,y, color='purple')
plt.plot(X, lin_regressor.predict(X), color=('black'))
plt.title('truth or Bluff(Linear Regression)')
plt.xlabel('Position level')
plt.ylabel('salary')
plt.show()
```

##**plotting prediction of polynomial regression, actually overfitting it by changing the degree to 2 in above PR code**

```python
plt.scatter(X,y, color='purple')
plt.plot(X, lin_reg2.predict(X_poly), color=('black'))
plt.title('truth or Bluff(Polynomial Regression)')
plt.xlabel('Position level')
plt.ylabel('salary')
plt.show()
```

##**predicting for our linear regression model**

```python
lin_regressor.predict([[6.5]])
```

##**predicting for poly regression model**

```python
lin_reg2.predict(poly_regressor.fit_transform([[6.5]]))
```
